```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.metrics import precision_recall_curve, f1_score, roc_curve
from imblearn.over_sampling import SMOTE
from catboost import CatBoostClassifier, Pool
import shap
import warnings
warnings.filterwarnings('ignore')

# ==============================
# 1. LOAD DATA
# ==============================

data = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# ==============================
# 2. CLEANING & TYPE FIXES
# ==============================

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
if data['TotalCharges'].isnull().sum() > 0:
    data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)

# ==============================
# 3. FEATURE ENGINEERING
# ==============================
# tenure bucket
data['tenure_bucket'] = pd.cut(
    data['tenure'], bins=[0, 12, 24, 48, np.inf], labels=['0-12', '13-24', '25-48', '49+'], include_lowest=True
)

# service binaries + services_count
service_cols = [
    'PhoneService','OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies'
]

for s in service_cols:
    data[s + '_bin'] = data[s].replace({
        'No': 0, 'Yes': 1,
        'No internet service': 0,
        'No phone service': 0
    }).astype(int)

data['services_count'] = data[[s + '_bin' for s in service_cols]].sum(axis=1)

# contract-tenure interaction
data['contract_len'] = data['Contract'].map({
    'Month-to-month': 1,
    'One year': 12,
    'Two year': 24
})
data['contract_tenure_ratio'] = data['tenure'] / (data['contract_len'] + 1e-6)

# ==============================
# 4. TARGET VARIABLE
# ==============================

y = (data['Churn'] == 'Yes').astype(int)

# Drop non-features
X = data.drop(columns=['customerID', 'Churn'])

# Identify categorical features for CatBoost
cat_features = [i for i, col in enumerate(X.columns) if X[col].dtype == 'object' or X[col].dtype == 'category']

# ==============================
# 5. SPLIT INTO TRAIN/VAL/TEST (70/15/15)
# ==============================

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.17647, stratify=y_temp, random_state=42
)

# ==============================
# 6. OPTIONAL — SMOTE FOR NON-TREE MODELS (NOTUSED BY CATBOOST)
# If you want to test LR/SVM:
# sm = SMOTE(random_state=42)
# X_train_res, y_train_res = sm.fit_resample(X_train.select_dtypes(include=[np.number]), y_train)

# ==============================
# 7. CATBOOST MODEL
# ==============================

train_pool = Pool(X_train, y_train, cat_features=cat_features)
val_pool = Pool(X_val, y_val, cat_features=cat_features)

tuned_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.05,
    depth=6,
    eval_metric='F1',
    random_seed=42,
    early_stopping_rounds=50,
    verbose=100
)

tuned_model.fit(train_pool, eval_set=val_pool)

# ==============================
# 8. EVALUATION
# ==============================

probs = tuned_model.predict_proba(X_test)[:, 1]
preds_05 = (probs > 0.5).astype(int)

print("\nClassification Report (0.5 Threshold):\n", classification_report(y_test, preds_05))
print("ROC AUC:", roc_auc_score(y_test, probs))

# ==============================
# 9. CONFUSION MATRIX PLOT
# ==============================

cm = confusion_matrix(y_test, preds_05)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Threshold = 0.5)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300)
plt.close()

# ==============================
# 10. TRAINING CURVE
# ==============================

results = tuned_model.get_evals_result()
train_f1 = results['learn']['F1']
val_f1 = results['validation']['F1']

plt.figure(figsize=(8,5))
plt.plot(train_f1, label='Training F1')
plt.plot(val_f1, label='Validation F1')
plt.xlabel('Iteration')
plt.ylabel('F1 Score')
plt.title('Training Curve (CatBoost F1)')
plt.legend()
plt.tight_layout()
plt.savefig('training_curve.png', dpi=300)
plt.close()

# ==============================
# 11. ROC CURVE
# ==============================

fpr, tpr, _ = roc_curve(y_test, probs)
plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label='ROC')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.tight_layout()
plt.savefig('roc_curve.png', dpi=300)
plt.close()

# ==============================
# 12. PRECISION-RECALL CURVE
# ==============================

precision, recall, thresholds = precision_recall_curve(y_test, probs)
plt.figure(figsize=(7,5))
plt.plot(recall, precision, label='PR Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.tight_layout()
plt.savefig('pr_curve.png', dpi=300)
plt.close()

# ==============================
# 13. THRESHOLD TUNING (F1)
# ==============================

f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)
best_idx = np.argmax(f1_scores)
best_thresh = thresholds[best_idx]
print("Best threshold (by F1):", best_thresh)

best_preds = (probs > best_thresh).astype(int)
print("\nClassification Report (Best Threshold):\n", classification_report(y_test, best_preds))

# ==============================
# 14. SHAP INTERPRETATION
# ==============================

explainer = shap.TreeExplainer(tuned_model)
shap_values = explainer.shap_values(X_val.head(500))  # sample for speed

plt.figure(figsize=(10,6))
shap.summary_plot(shap_values, X_val.head(500), show=False)
plt.tight_layout()
plt.savefig('shap_summary.png', dpi=300)
plt.close()

print("\nEverything completed successfully. Figures saved:")
print("- training_curve.png")
print("- roc_curve.png")
print("- pr_curve.png")
print("- confusion_matrix.png")
print("- shap_summary.png")
```
